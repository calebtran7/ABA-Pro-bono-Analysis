```python
import pandas as pd
```

```python
clients_df = pd.read_csv("clients.csv")
attorneys_df = pd.read_csv("attorneys.csv")
categories_df = pd.read_csv("categories.csv")
questions_df = pd.read_csv("questions.csv")
```

```python
clients_df.head()
```

```python
attorneys_df.head()
```

```python
categories_df["Category"].unique()
```

```python
questions_df.head()
questions_df.rename(columns = {'AskedByClientUno':'ClientUno'}, inplace = True)
questions_df.rename(columns = {'TakenByAttorneyUno':'AttorneyUno'}, inplace = True)
questions_df.head()
```

```python
attorneyTimeEntries_df = pd.read_csv("attorneytimeentries.csv")
cols_keep = ["AttorneyUno", "Hours"]
attorneyTimeEntries_df = attorneyTimeEntries_df[cols_keep]
attorneyTimeEntries_df.head()
```

```python
merged = pd.merge(clients_df, questions_df, 
                   on='ClientUno', 
                   how='inner')
merged.head()
```

```python
merged.rename(columns = {'TakenByAttorneyUno':'AttorneyUno'}, inplace = True)
```

```python
innerMerge = pd.merge(merged, attorneys_df, on='AttorneyUno', how='inner')
innerMerge.reset_index(drop=True)
```

```python
innerMerge["EthnicIdentity"].value_counts().head(10)
```

```python
innerMerge.head()
innerMerge.to_csv("questionsMatched.csv")
```

```python
consumerFinancialQuestions = innerMerge[innerMerge.Category == "Consumer Financial Questions"]
cols_to_show = ["Category", "County_x", "StateName_x", "EthnicIdentity", "Age", "Gender", "MaritalStatus", "AnnualIncome", "NumberInHousehold"]
consumerFinancialQuestions = consumerFinancialQuestions[cols_to_show]
```

```python
consumerFinancialQuestions['count'] = 1
consumerFinancialQuestions.head()
consumerFinancialQuestions.to_csv("consumerFinancialQuestions.csv")
```

```python
consumerFinancialQuestions.groupby('EthnicIdentity').sum().sort_values(by = "count", ascending = False).head(10)
```
